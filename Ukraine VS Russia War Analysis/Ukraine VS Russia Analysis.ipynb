```python
equipment=pd.read_csv('russia_losses_equipment.csv')
personnel=pd.read_csv('russia_losses_personnel.csv')
```

Let's **Merge** these two dataset

```python
equipment[['personnel','POW']]=personnel[['personnel','POW']]
equipment.head()
```

```python
df=equipment.copy()
df.shape
```

```python
df.dtypes
```

```python
# Cleaning the 'great losses direction' feature

great_loss = equipment[["greatest losses direction"]]
great_loss["greatest losses direction"].replace(np.nan,"NA",inplace=True)
great_loss = great_loss[great_loss["greatest losses direction"].isin(["NA"])==False]
great_loss["greatest losses direction"] = great_loss["greatest losses direction"].str.split(", | , | ,| and ")
great_loss = great_loss.explode("greatest losses direction")
great_loss.replace(to_replace = ["Zaporizhzhya","Slov'yansʹk"], value = ["Zaporizhzhia","Sloviansk"],inplace=True)

print(f"The shape is : {great_loss.shape}\n")
print(f'Unique great losses directions are :\n\n{great_loss["greatest losses direction"].unique()}')
print(f'\nTotal unique great losses directions is {len(great_loss["greatest losses direction"].unique())}')
print(f'\nOccurance of unique losses direction :\n\n{great_loss["greatest losses direction"].value_counts()}')
```

```python
plt.figure(figsize=(14,8))
sns.countplot(great_loss['greatest losses direction'].sort_values(ascending=True))
plt.title('DIrection wise Occurance',fontsize=15)
plt.xlabel('Greatest Losses Direction')
plt.ylabel('Counts')
plt.xticks(rotation=45)
plt.show()
```

* Let's drop **greatest losses direction** column, because i don't need these for further     
* merge **"military auto","fuel tank"** and **vehicle and fuel tanks** columns into **vehicle and fuel tanks** 

```python
war_data=df[:]
war_data['vehicle and fuel tanks']=war_data[['military auto','fuel tank','vehicles and fuel tanks']].sum(axis=1)
war_data.drop(['date','military auto','fuel tank','greatest losses direction'],axis=1,inplace=True)
```

```python
# Now Handling with NaN values
war_data.isna().sum().to_frame().T
```

So we have some null values for **"Special Equipment","mobile SRBM system","vehicle nd fuel tanks"** and **"Cruise missiles**. Lets replace them with **0**.

```python
war_data.fillna(0,inplace=True)
```

```python
war_data.isna().sum().to_frame().T
```

**Note:**     
* We can see all the values of the dataset are recorded in column-wise **cumulative sum format**.Let's create a new dataset that in day-wise recorded format.

### Person Data

```python
person_data=war_data[['personnel','POW']].diff().fillna(war_data[['personnel','POW']])
person_data.head(10).T
```

```python
print("Total number of dead Invader : {}".format(person_data['personnel'].sum()))
print("Total amount of prisoners: {}".format(person_data['POW'].sum()))
```

```python
plt.figure(figsize=(9,9))

labels='Dead Personnel','Prosoners'
size=0.7

wedges,texts,autexts=plt.pie([person_data['personnel'].sum(),person_data['POW'].sum()],explode={0,0.1},
                             textprops=dict(size=15,color='white'),autopct='%.2f%%',pctdistance=0.80,radius=1,shadow=True,
                            wedgeprops=dict(width=size,edgecolor='white',linewidth=5),colors=["#0f4c5c","#FFC300"])
plt.legend(wedges,labels,title='Category',loc='center left',bbox_to_anchor=(1,0,0.5,1))
plt.title('Density of Invaders',fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.histplot(person_data['personnel'],kde=True,bins='auto',alpha=1,fill=True)
plt.title('Personnel Death Quantity Distribution',fontsize=20)
plt.xlabel('Personnel Amount')
plt.ylabel('Count')
plt.show()
```

# Exploring AIR Equipment 

```python
air_data=war_data[['aircraft','helicopter','drone','mobile SRBM system']].diff().fillna(war_data[['aircraft','helicopter','drone','mobile SRBM system']])
air_data['day']=war_data['day']
air_data.head()
```

```python
_,ax=plt.subplots(2,2,figsize=(20,10),sharex=True)
plt.tight_layout(pad=5.0)

sns.lineplot(x='day',y='aircraft',data=air_data,ax=ax[0,0],color='red')
ax[0,0].set_title('Aircraft Destruction',fontsize=20)
ax[0,0].set_xlabel('Days')
ax[0,0].set_ylabel('Quantity')

sns.lineplot(x='day',y='drone',data=air_data,ax=ax[0,1],color='blue')
ax[0,1].set_title('Drone Destruction',fontsize=20)
ax[0,1].set_xlabel('Days')
ax[0,1].set_ylabel('Quantity')

sns.lineplot(x='day',y='helicopter',data=air_data,ax=ax[1,0],color='green')
ax[1,0].set_title('Helicopter Destruction',fontsize=20)
ax[1,0].set_xlabel('Days')
ax[1,0].set_ylabel('Quantity')

sns.lineplot(x='day',y='mobile SRBM system',data=air_data,ax=ax[1,1],color='orange')
ax[1,1].set_title('Mobile SRBM system Destruction',fontsize=20)
ax[1,1].set_xlabel('Days')
ax[1,1].set_ylabel('Quantity')
plt.show()
```

```python
plt.figure(figsize=(12,12))
labels=['Aircraft','Helicopter','Drone']
size=0.7

wedges,texts,autotexts=plt.pie([air_data['aircraft'].sum(),air_data['helicopter'].sum(),air_data['drone'].sum()],
                               explode=(0.1,0,0),textprops=dict(size=20,color='white'),autopct='%.2f%%',pctdistance=0.65,
                              colors=["#FFC300","#0f4c5c","#3B3F46"],shadow=True,wedgeprops=dict(width=size,edgecolor='white',
                              linewidth=5),startangle=-30)
plt.legend(wedges,labels,title='Category',bbox_to_anchor=(1,0,0.5,1))
plt.title('Air Distruction',fontsize=20)
plt.show()
```

```python
_, axs = plt.subplots(1,2,figsize=(20,5),sharex=True)
plt.tight_layout(pad=4.0)

sns.lineplot(x="day",y="naval ship",data=water_data,ax=axs[0],color='red')
axs[0].set_title("Naval Ship Destruction",fontsize=20)
axs[0].set_xlabel("Days")
axs[0].set_ylabel("Quantity")

sns.lineplot(x="day",y="cruise missiles",data=water_data,ax=axs[1],color='green')
axs[1].set_title("Cruise Missiles Destruction",fontsize=20)
axs[1].set_xlabel("Days")
axs[1].set_ylabel("Quantity")

sns.despine(left=True, bottom=True)
plt.show()
```

```python
plt.subplots(figsize=(12, 12))

labels = "Naval Ship","Cruise Missiles"
size = 0.7

wedges, texts, autotexts = plt.pie([water_data["naval ship"].sum(),
                                    water_data["cruise missiles"].sum()],
                                    explode = (0.1,0),
                                    textprops=dict(size= 20, color= "white"),
                                    autopct="%.2f%%", 
                                    pctdistance = 0.65,
                                    radius=.9, 
                                    colors = ["#FFC300","#3B3F46"], 
                                    # labels = labels, # labeldistance = 1.05,
                                    shadow = True,
                                    wedgeprops=dict(width = size, edgecolor = "white", 
                                    linewidth = 5),
                                    startangle = -15)
plt.legend(wedges, labels, title="Category",loc="center left",bbox_to_anchor=(1, 0, 0.5, 1))
plt.title("\nWater Destruction",fontsize=20)
plt.show()
```

# Exploring Field Equipments           
<!--  -->          
As **Vehicle and fuel tanks** has negative values,let's set it to zero

```python
field_data = war_data[["tank","APC","MRL","field artillery","anti-aircraft warfare","vehicles and fuel tanks"]].diff().fillna(war_data[["tank","APC","MRL","field artillery","anti-aircraft warfare","vehicles and fuel tanks"]])
field_data[field_data["vehicles and fuel tanks"] < 0]["vehicles and fuel tanks"] = 0
field_data[["day"]] = war_data[["day"]]
field_data.head()
```

```python
_, axs = plt.subplots(2,3,figsize=(25,10),sharex=True)
plt.tight_layout(pad=4.0)

sns.lineplot(x="day",y="tank",data=field_data,ax=axs[0,0],color='red')
axs[0,0].set_title("\nTank Destruction\n",fontsize=20)
axs[0,0].set_xlabel("Days")
axs[0,0].set_ylabel("Quantity")

sns.lineplot(x="day",y="APC",data=field_data,ax=axs[0,1],color='grey')
axs[0,1].set_title("\nAPC Destruction\n",fontsize=20)
axs[0,1].set_xlabel("Days")
axs[0,1].set_ylabel("Quantity")

sns.lineplot(x="day",y="vehicles and fuel tanks",data=field_data,ax=axs[0,2],color='orange')
axs[0,2].set_title("\nVehicles and Fuel Tanks Destruction\n",fontsize=20)
axs[0,2].set_xlabel("Days")
axs[0,2].set_ylabel("Quantity")

sns.lineplot(x="day",y="MRL",data=field_data,ax=axs[1,0],color='green')
axs[1,0].set_title("\nMRL Destruction\n",fontsize=20)
axs[1,0].set_xlabel("Days")
axs[1,0].set_ylabel("Quantity")

sns.lineplot(x="day",y="anti-aircraft warfare",data=field_data,ax=axs[1,1],color='yellow')
axs[1,1].set_title("\nAnti-Aircraft Warfare Destruction\n",fontsize=20)
axs[1,1].set_xlabel("Days")
axs[1,1].set_ylabel("Quantity")

sns.lineplot(x="day",y="field artillery",data=field_data,ax=axs[1,2],color='pink')
axs[1,2].set_title("\nField Artillery Destruction\n",fontsize=20)
axs[1,2].set_xlabel("Days")
axs[1,2].set_ylabel("Quantity")

sns.despine(left=True, bottom=True)
plt.show()
```

```python
plt.subplots(figsize=(12, 12))

labels = "Tank","MRL","Anti Aircraft Warfare","APC","Field Artillery","Vehicles And Fuel Tanks"
size = 0.7

wedges, texts, autotexts = plt.pie([field_data["tank"].sum(),
                                    field_data["MRL"].sum(),
                                    field_data["anti-aircraft warfare"].sum(),
                                    field_data["APC"].sum(),
                                    field_data["field artillery"].sum(),
                                    field_data["vehicles and fuel tanks"].sum()],
                                    explode = (0,0,0,0,0.1,0),
                                    textprops=dict(size= 20, color= "white"),
                                    autopct="%.2f%%", 
                                    pctdistance = 0.8,
                                    radius=1, 
                                    colors = ["#FFC300","#9e2a2b","#99582a","#3B3F46","#d1ac00","#1d7874"], 
                                    # labels = labels, # labeldistance = 1.05,
                                    shadow = True,
                                    wedgeprops=dict(width = size, edgecolor = "white", 
                                    linewidth = 5),
                                    startangle = 110)

plt.legend(wedges, labels, title="Category",loc="center left",bbox_to_anchor=(1, 0, 0.5, 1))
plt.title("\nField Destruction",fontsize=20)
plt.show()
```

# Exploring Day-Wise Desctruction

```python
_, axs = plt.subplots(figsize=(20,10))
plt.tight_layout(pad=4.0)

sns.lineplot(x=air_data["day"],y=air_data[["aircraft","helicopter","drone","mobile SRBM system"]].sum(axis=1),ax=axs)
sns.lineplot(x=water_data["day"],y=water_data[["naval ship","cruise missiles"]].sum(axis=1),ax=axs)
sns.lineplot(x=field_data["day"],y=field_data[["tank","APC","MRL","field artillery","anti-aircraft warfare","vehicles and fuel tanks"]].sum(axis=1),ax=axs)

axs.set_title("\nWar Equipment Destruction\n",fontsize=20)
axs.set_xlabel("Days")
axs.set_ylabel("Quantity")
axs.legend(["Air Equipment", "Water Equipment", "Field Equipment"])
sns.despine(left=True, bottom=True)

plt.show()
```

# Exploring Cumulative Destruction

```python
_, axs = plt.subplots(figsize=(20,10))
plt.tight_layout(pad=4.0)

sns.lineplot(x=air_data["day"],y=air_data[["aircraft","helicopter","drone","mobile SRBM system"]].sum(axis=1).cumsum(axis=0),ax=axs)
sns.lineplot(x=water_data["day"],y=water_data[["naval ship","cruise missiles"]].sum(axis=1).cumsum(axis=0),ax=axs)
sns.lineplot(x=field_data["day"],y=field_data[["tank","APC","MRL","field artillery","anti-aircraft warfare","vehicles and fuel tanks"]].sum(axis=1).cumsum(axis=0),ax=axs)

axs.set_title("\nWar Equipment Destruction\n",fontsize=20)
axs.set_xlabel("Days")
axs.set_ylabel("Quantity")
axs.legend(["Air Equipment", "Water Equipment", "Field Equipment"])
sns.despine(left=True, bottom=True)

plt.show()
```
