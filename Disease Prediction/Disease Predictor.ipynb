# Model Building 

```python
# splitting the dataset into train and test
from sklearn.model_selection import train_test_split
X=df.drop('prognosis',axis=1)
y=df['prognosis']

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import ShuffleSplit,GridSearchCV
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
# Parameter tuning for decision tree
def tuning(X,y):
    models={'decisiontree':{'model':DecisionTreeClassifier(splitter='best'),'parameters':{'criterion':['gini','entropy'],
                                                                                         'max_depth':[5,10,12,16]}}}
    score=[]
    cv_shuffle=ShuffleSplit(n_splits=5,test_size=0.1,random_state=0)
    
    for model_name,model_params in models.items():
        gs=GridSearchCV(model_params['model'],model_params['parameters'],cv=cv_shuffle,return_train_score=False)
        gs.fit(X,y)
        score.append({'model':model_name,'best_parameters':gs.best_params_,'score':gs.best_score_})
    return pd.DataFrame(score,columns=['model','best_parameters','score'])
tuning(X_train,y_train)
        
```

```python
# fitting the model
dt=DecisionTreeClassifier(criterion='entropy',max_depth=16)
dt.fit(X_train,y_train)
pred=dt.predict(X_test)
print('test accuracy: {}'.format(accuracy_score(pred,y_test)))
```
